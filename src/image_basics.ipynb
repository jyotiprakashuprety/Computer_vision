```python
import numpy as np
from matplotlib import pyplot as plt
from skimage import img_as_ubyte
```

We can just randomly generate image, as it is just a array of numbers with pixels intensity value

```python
random_image = np.random.random([256,256])
plt.imshow(random_image)
```

Being a numpy array, any numpy operatioin could perform on image

```python
lower_intensity_image = img_as_ubyte(random_image * 0.5)
plt.imshow(lower_intensity_image)
```

```python
print(lower_intensity_image.max(), lower_intensity_image.min())
print(random_image.max(), random_image.min())
```
